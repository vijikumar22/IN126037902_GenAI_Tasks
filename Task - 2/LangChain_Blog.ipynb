3. Basic LLM Call

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-3.5-turbo")

response = llm.invoke("Explain LangChain in simple terms")
print(response.content)
```

4. Prompt Template

```python
from langchain.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

formatted_prompt = prompt.format(topic="Machine Learning")
print(formatted_prompt)
```

5. Chain Example

```python
from langchain.chains import LLMChain

chain = LLMChain(llm=llm, prompt=prompt)

result = chain.run("Artificial Intelligence")
print(result)
```

6. Memory Example

```python
from langchain.memory import ConversationBufferMemory

memory = ConversationBufferMemory()

memory.save_context({"input": "Hi"}, {"output": "Hello!"})
memory.save_context({"input": "How are you?"}, {"output": "I'm fine!"})

print(memory.load_memory_variables({}))
```

7. Tool Creation

```python
from langchain.tools import tool

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b

print(multiply.run({"a": 5, "b": 4}))
```

8. Agent with Tool

```python
from langchain.agents import initialize_agent, AgentType
from langchain.tools import Tool

def calculator(input):
    return eval(input)

tools = [
    Tool(
        name="Calculator",
        func=calculator,
        description="Useful for math calculations"
    )
]

agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

result = agent.run("What is 25 * 4?")
print(result)
```

9. Simple Document Loader

```python
from langchain.document_loaders import TextLoader

# Create a sample file first
with open("sample.txt", "w") as f:
    f.write("LangChain is a framework for building LLM applications.")

loader = TextLoader("sample.txt")
documents = loader.load()

print(documents)
```

10. Final Combined Flow Demo

```python
# Full pipeline demo

query = "Explain Artificial Intelligence"

formatted = prompt.format(topic=query)

response = llm.invoke(formatted)

print("Final Output:\n", response.content)
```
